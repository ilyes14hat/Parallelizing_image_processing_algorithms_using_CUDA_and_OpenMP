# <font color="cyan">**Results of comparison between Sequential and OpenMP parallel algorithm**</font>

## <font color="lime"> 0. Graph function</font>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Same Image Size

```python
def graph_openmp(table1,table2,title):
  plt.figure(figsize = (15,7))
  plt.plot(table1, label = 'Sequential')
  plt.plot(table2, label = 'OpenMP')
  plt.xlabel('Number of test',fontsize=14)
  plt.ylabel('Time in Seconds',fontsize=14)
  plt.title(title, fontsize=20)
  plt.legend()
  plt.show()
```

### Different Image Size

```python
def graph_openmp_diff(table1,title):
  plt.figure(figsize = (15,7))
  plt.plot(["1024x1024","2048x2048","4096x4096","8192x8192"],table1)
  plt.xlabel('Image Size',fontsize=14)
  plt.ylabel('Accelartion',fontsize=14)
  plt.ylim([1, 4])
  plt.title(title, fontsize=20)
  plt.show()
```

## <font color="lime">1. Invert Function</font>

### <font color="gold">1.1 Same Image Size

```python
Sequential_Time_8192_pixels_invert_1 = [0.110268,0.115142,0.114161,0.114315,0.114826,0.116787,0.110747,0.103829,0.127779,0.122435]
Parallel_OpenMP_Time_8192_pixels_invert = [0.036590,0.045620,0.038507,0.032874,0.035082,0.042687,0.033572,0.044204,0.038298,0.049837]

graph_openmp(Sequential_Time_8192_pixels_invert_1, Parallel_OpenMP_Time_8192_pixels_invert,'Comparison on the invert function on a 8192 x 8192 px image')
```

### <font color="gold">1.2 Different Image Sizes

```python
Sequential_Time_diff_size_invert_1 = np.array([0.110268,0.223837,0.445621,0.891312])
Parallel_OpenMP_Time_diff_size_invert = np.array([0.036590,0.069218,0.133129,0.257134])
acc_inver_openmp = Sequential_Time_diff_size_invert_1/Parallel_OpenMP_Time_diff_size_invert

graph_openmp_diff(acc_inver_openmp,'Comparison on the invert function on different image sizes')
```

### Same Image Size

```python
import matplotlib.pyplot as plt
import pandas as pd  
```

```python
def graph(table1,table2,title):
  plt.figure(figsize = (15,7))
  plt.plot(table1, label = 'Sequential')
  plt.plot(table2, label = 'CUDA')
  plt.xlabel('Number of test',fontsize=14)
  plt.ylabel('Time in Seconds',fontsize=14)
  plt.title(title, fontsize=20)
  plt.legend()
  plt.show()
```
